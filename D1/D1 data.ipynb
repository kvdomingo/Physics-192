```python
import numpy as np
import matplotlib.pyplot as mp
import matplotlib.ticker as tick
import scipy.optimize as opt
import scipy.stats as stat
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.style.use("seaborn")
mp.rcParams["font.family"] = "serif"
mp.rcParams["text.usetex"] = True
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
```

```python
coin = np.loadtxt("coin.txt")
```

```python
def decayfit(t, A, tau):
    return A*np.exp(-t/tau)
```

```python
popt, pcov = opt.curve_fit(decayfit, np.arange(0, len(coin), 1), np.log(coin))

mp.plot(range(len(coin)), np.log(coin), "o", label="data")

x = np.linspace(0, len(coin), 100)
mp.plot(x, decayfit(x, *popt), label=r"data fit: $A_0 = %.2f$, $\tau = %.2f$"%(tuple(popt)))
mp.plot(x, np.log(100)*np.exp(-1/2*x), "--", label=r"theoretical")

mp.legend()
mp.show()

print("time constant (tau) = %.2f min"%(popt[-1]))
print("initial count (A0) = %.2f counts"%(popt[0]))
print("decay constant (lambda) = %.2f per min"%(1/popt[-1]))
print("half life (t_1/2) = %.2f min"%(np.log(2)*popt[-1]))
```
