```python
import numpy as np
import matplotlib.pyplot as mp
import scipy.stats as stat
import uncertainties as uc
import uncertainties.umath as um
import uncertainties.unumpy as unp
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["font.family"] = "serif"
mp.rcParams["text.usetex"] = True
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
```

```python
# LEFT reading v RIGHT reading
err = 1/60
central_left = unp.uarray([np.radians(177 + 6/60)], err)
central_right = unp.uarray([np.radians(356.5 + 6/60)], err)

# violet blue cyan red
n = np.array([6, 5, 4, 3])

theta_left1 = unp.uarray([np.radians(184 + 8/60), np.radians(184.5 + 8/60), np.radians(185 + 10/60), np.radians(187 + 5/60)], err) - central_left
theta_right1 = central_right - unp.uarray([np.radians(4 + 9/60), np.radians(4.5 + 5/60), np.radians(5 + 13/60), np.radians(9 + 11/60)], err)

theta_left2 = central_left - unp.uarray([np.radians(164 + 6/60), np.radians(161 + 8/60), np.radians(158 + 9/60), np.radians(154 + 9/60)], err)
theta_right2 = unp.uarray([np.radians(342 + 3/60), np.radians(344 + 7/60), np.radians(340 + 8/60), np.radians(334 + 12/60)], err) - central_right
```

```python
a = 1/300 * 1e-3

lambda_left1 = a*unp.sin(theta_left1)/1
lambda_left2 = a*unp.sin(theta_left2)/2
lambda_right1 = a*unp.sin(theta_right1)/1
lambda_right2 = a*unp.sin(theta_right2)/2

x = 1/2**2 - 1/n**2
yl1 = 1/lambda_left1
yl2 = 1/lambda_left2
yr1 = 1/lambda_right1
yr2 = 1/lambda_right2
```

```python
m1, b1, R1, p1, sd1 = stat.linregress(x, unp.nominal_values(yl1))
m2, b2, R2, p2, sd2 = stat.linregress(x, unp.nominal_values(yl2))

mp.plot(x, m1*x+b1, "b--", label=r"fit: $m = %.2f \times 10^7$, $b = %.2f \times 10^5$, $R^2 = %.3f$"%(m1*1e-7, b1*1e-5, R1**2))
mp.plot(x, (m2*x+b2), "r--", label=r"fit: $m = %.2f \times 10^7$, $b = %.2f \times 10^5$, $R^2 = %.3f$"%((m2)*1e-7, b2*1e-5, R2**2))

mp.errorbar(x, unp.nominal_values(yl1), unp.std_devs(yl1), 0, "bo", label="left reading, 1st order")
mp.errorbar(x, (unp.nominal_values(yl2)), unp.std_devs(yl2), 0, "ro", label="right reading, 1st order")

mp.xlabel(r"$2^{-2} - n^{-2}$")
mp.ylabel(r"$\lambda^{-1}$")
mp.legend()

mp.savefig("rydberg_1.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
m3, b3, R3, p3, sd3 = stat.linregress(x, unp.nominal_values(yr1))
m4, b4, R4, p4, sd4 = stat.linregress(x, unp.nominal_values(yr2))

mp.plot(x, (m3*x+b3), "b--", label=r"fit: $m = %.2f \times 10^7$, $b = %.2f \times 10^5$, $R^2 = %.3f$"%((m3)*1e-7, b3*1e-5, R3**2))
mp.plot(x, (m4*x+b4), "r--", label=r"fit: $m = %.2f \times 10^7$, $b = %.2f \times 10^5$, $R^2 = %.3f$"%((m4)*1e-7, b4*1e-5, R4**2))

mp.errorbar(x, (unp.nominal_values(yr1)), unp.std_devs(yr1), 0, "bo", label="left reading, 2nd order")
mp.errorbar(x, (unp.nominal_values(yr2)), unp.std_devs(yr2), 0, "ro", label="right reading, 2nd order")

mp.xlabel(r"$2^{-2} - n^{-2}$")
mp.ylabel(r"$\lambda^{-1}$")
mp.legend()

mp.savefig("rydberg_2.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
R_theo = 1.097e7
err = abs(R_theo - m1)/R_theo * 100
print("Experimental R_H (left reading, 1st order) = {0:.2f}E7\n{1:.2f}% deviation".format(m1*1e-7, err))
```

```python
err = abs(R_theo - abs(m2))/R_theo * 100
err, abs(m2)
print("Experimental R_H (right reading, 1st order) = {0:.2f}E7\n{1:.2f}% deviation".format(abs(m2)*1e-7, err))
```

```python
err = abs(R_theo - abs(m3))/R_theo * 100
err, abs(m3)
print("Experimental R_H (left reading, 2nd order) = {0:.2f}E7\n{1:.2f}% deviation".format(abs(m3)*1e-7, err))
```
