```python
import numpy as np
import matplotlib.pyplot as mp
from astropy.constants import g0; g0 = g0.value
import scipy.stats as stat
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
mp.rcParams["text.usetex"] = True
mp.rcParams["font.family"] = "serif"
```

```python
n = np.array([-5, -4, -3, -2, -1, 1, 2, 3, 4, 5])
f = np.array([39, 40, 41, 37, 31, 27, 33, 36, 36, 35])/60
R = 0.973 #total length
sep = 0.092 #edge-to-edge separation
d = 0.017 #diameter
inity = 0.05 #initial horizontal displacement
T = 1/f
center = R/2
L = (n*(sep - d))
```

```python
mp.plot(L, T, "ko")
mp.grid()
mp.xlabel("distance from CG")
mp.ylabel("T")
mp.show()
```

```python
x = abs(L)*T**2
y = abs(L)**2
m, b, R, p, sd = stat.linregress(x, y)

mp.plot(x, m*x+b, "k--", label=r"$m = %.2f$, $b = %.2f$, $R^2 = %.3f$"%(m, b, R**2))
mp.plot(x, y, "ko", label="data")
mp.grid()
mp.xlabel("$LT^2$")
mp.ylabel("$L^2$")
mp.legend()
mp.show()
```
