```python
import numpy as np
import matplotlib.pyplot as mp
import scipy.stats as stat
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["text.usetex"] = True
mp.rcParams["font.family"] = "serif"
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
```

```python
t, Vin, Vout = np.genfromtxt("openend2.csv", delimiter=",", skip_header=2, usecols=(0,1,2), unpack=True)
t += abs(t.min())
L = 49
c = 3e8
```

```python
mp.plot(t, Vin, "y-", label=r"$V_{in}$")
mp.plot(t, Vout, "c-", label=r"$V_{out}$")
mp.legend()
mp.grid(True)
mp.xlabel("time, s")
mp.ylabel("amplitude, V")
mp.show()
```

```python
c_expt = 2*L/abs(t[np.argmax(Vout)] - t[np.argmax(Vin)])
c_expt
c_expt/2
```
