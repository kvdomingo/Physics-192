```python
import numpy as np
import matplotlib.pyplot as mp
from astropy.constants import g0; g0 = g0.value
import uncertainties as uc
import uncertainties.unumpy as unp
import scipy.optimize as opt
import scipy.stats as stat
import os
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["text.usetex"] = True
mp.rcParams["font.family"] = "serif"
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
```

```python
unp_n = unp.nominal_values
unp_sd = unp.std_devs
```

```python
def g(t, a, f, phi, b, c):
    return a * np.sin(2*np.pi*f*t - phi) * np.exp(-b*t) + c
```

```python
xerr = 1
merr = 4
x = unp.uarray([6, 8.2, 10, 11.4, 12.1, 13.1, 14.7, 18, 35.4], xerr) * 1e-2
m = unp.uarray([0, 20, 47.6, 67.6, 100, 120, 147.6, 200, 478.4], merr) * 1e-3
F = m*g0
```

```python
k, b, R, p, sd = stat.linregress(unp_n(x), unp_n(F))

mp.plot(unp_n(x), k*unp_n(x)+b, "k--", label=r"$k = %.2f$, $F_0 = %.2f$, $R^2 = %.3f$"%(k, b, R**2))
mp.errorbar(unp_n(x), unp_n(F), None, unp_sd(x), "ko", label="data")
mp.grid(True)
mp.legend()
mp.xlabel("displacement, m")
mp.ylabel("weight, N")

#mp.savefig("springk.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
flist = np.arange(0.5, 2.1, 0.1)
Alist = []
philist = []

for _, __, files in os.walk("."):
    for file in files:
        if file.endswith(".txt"):
            t, x, v, a = np.genfromtxt(file, skip_header=7).T
            popt = opt.curve_fit(g, t, x, bounds=([0, 0, -np.inf, -np.inf, -np.inf], np.inf))[0]
            Alist.append(x.max())
            philist.append(popt[2])
            
Alist = np.array(Alist)
philist = np.array(philist)
```

```python
mp.plot(flist, Alist, "ko")
mp.xlabel("frequency, Hz")
mp.ylabel("displacment, m")
mp.grid(True)

#mp.savefig("resonance.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
mp.plot(flist, philist, "ko")
mp.xlabel("frequency, Hz")
mp.ylabel("phase, rad")
mp.grid(True)

#mp.savefig("phase.png", dpi=300, bbox_inches="tight")
mp.show()
```
