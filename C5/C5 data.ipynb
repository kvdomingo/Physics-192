```python
import numpy as np
import matplotlib.pyplot as mp
import scipy.stats as stat
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["text.usetex"] = True
mp.rcParams["font.family"] = "serif"
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
```

```python
PS, LQ1, LQ2, LQ3, LQave, E1, E2, E3, Eave, FE, ABS = np.genfromtxt("corr.csv", skip_header=1, 
                                                                    usecols=range(11), unpack=True, delimiter=",")
datx = np.array([LQ1, LQ2, LQ3, LQave, LQave, Eave, ABS, ABS, ABS, ABS, ABS, ABS, ABS])
labelx = np.array(["LQ1", "LQ2", "LQ3", "LQave", "LQave", "Eave", "ABS", "ABS", "ABS", "ABS", "ABS", "ABS", "ABS"])
daty = np.array([E1, E2, E3, Eave, FE, FE, LQ1, LQ2, LQ3, E1, E2, E3, FE])
labely = np.array(["E1", "E2", "E3", "Eave", "FE", "FE", "LQ1", "LQ2", "LQ3", "E1", "E2", "E3", "FE"])
assert(len(datx) == len(daty))
```

```python
for i in range(len(datx)):
    m, b, R = stat.linregress(datx[i], daty[i])[:3]
    x = np.linspace(datx[i].min(), datx[i].max())
    mp.plot(datx[i], daty[i], "ko")
    mp.plot(x, m*x+b, "k--")
    mp.grid(True)
    mp.xlabel(labelx[i])
    mp.ylabel(labely[i])
    mp.title(r"$R^2 = %.3f$"%(R**2))
    mp.show()
```
